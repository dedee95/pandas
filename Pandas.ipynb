# 6. Handling Missing and Categorical Data

## Handling missing data

```python
# create dummy dataset
data = {"A": [5, 10, 15, np.nan, 25],
        "B": [2, 4, np.nan, 8, 10],
        "C": [3, 6, 9, 12, 15]}
df = pd.DataFrame(data)
df
```

```python
# identifying not missing values
df.notnull()
```

```python
# identifying missing values
df.isnull()
```

```python
df.isna()
```

```python
# drop columns that contain missing values
df.dropna(axis="columns")
```

```python
# drop row that contain missing values
df.dropna(axis="index")
```

```python
# fill in missing value with zero
df.fillna(0)
```

```python
# fill in missing value with mean
df.fillna({"A": np.mean(df["A"]),
           "B": np.mean(df["B"]),
           "C": np.mean(df["C"])})
```

```python
# using ffill
df.fillna(method="ffill")
```

```python
# using bfill
df.fillna(method="bfill")
```

## Handling categorical data

```python
# creating dummy dataset
data = {"name": ["aziel", "martha", "felix", "nick", "yukio", "aida"],
        "gender": ["male", "female", "male", "male", "female", "female"]}
df = pd.DataFrame(data)
df
```

```python
# encoding nominal feature using .map()
encode = {"male": 0,
          "female": 1}
df["gender"] = df["gender"].map(encode)
df
```

```python
# encoding nominal feature using .replace()
df.replace(["male", "female"], [0, 1])
```

```python
# creating dummy dataset
data = {"name": ["aziel", "martha", "felix", "nick", "yukio", "aida"],
        "city": ["dallas", "chicago", "houston", "chicago", "houston", "dallas"]}
df = pd.DataFrame(data)
df
```

```python
# one hot encoding 
encode = pd.get_dummies(df["city"])
encode
```

```python
df = df.drop("city", axis="columns")
df = df.join(encode)
df
```

```python
# creating dummy dataset
data = {"name": ["aziel","martha","felix","nick","yukio"],
        "rating": ["frustrated","neutral","frustrated","satisfied","neutral"]}
df = pd.DataFrame(data)
df
```

```python
# encode ordinal feature
encode = {"frustrated": 0,
          "neutral": 1,
          "satisfied": 2}
df.replace(encode)
```

## Merge operator (inner, outer, left, and right)

```python
# creating dataframe 1
df1 = pd.DataFrame({"id": ["a", "b", "c", "d"],
                    "name": ["aziel", "martha", "felix", "nick"]})
df1
```

```python
# creating dataframe 2
df2 = pd.DataFrame({"id": ["c", "d", "e", "f"],
                    "gender": ["male", "female", "male", "male"]})
df2
```

```python
# inner join
pd.merge(df1, df2, on="id")
```

```python
# outer join
pd.merge(df1, df2, how="outer", on="id")
```

```python
# left join
pd.merge(df1, df2, how="left", on="id")
```

```python
# right join
pd.merge(df1, df2, how="right", on="id")
```

# 9. Create a Visualization from the DataFrame 

```python
df = pd.read_csv("Iris.csv")
df.drop(columns="Id", inplace=True)
df.head()
```

```python
# creating vertical bar plot
df["Species"].value_counts().plot(kind="bar")
```

```python
# creating horizontal bar plot
df["Species"].value_counts().plot(kind="barh")
```

```python
# create box plot 
grouped_df = df.groupby("Species") # grouping by species
grouped_df.get_group("Iris-virginica").plot(kind="box")
```

```python
# create kde plot 
grouped_df = df.groupby("Species") # grouping by species
grouped_df.get_group("Iris-virginica").plot(kind="kde")
```

```python
# enhance our visualization
df["Species"].value_counts().plot(kind="barh", 
                                  figsize=(8, 5), # figure size (inches)
                                  title="Horizontal Bar Plot", 
                                  legend=True,
                                  fontsize=14, 
                                  colormap="Dark2_r") 
```

```python
# change the backend 
pd.set_option("plotting.backend", "plotly") # set backend to plotly
```

```python
# horizontal bar plot with plotly
df["Species"].value_counts().plot(kind="barh")
```

```python
# box plot with plotly
pd.set_option("plotting.backend", "plotly") 
grouped_df = df.groupby("Species") 
grouped_df.get_group("Iris-virginica")["SepalWidthCm"].plot(kind="box")
```
